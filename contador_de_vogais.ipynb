```python
letras = str(input("digite alguma palavra para o bot contar a quantidade de vogais")).lower
contador = 0
vogais = ["a", "e", "i", "o", "u"]
for v in letras:
  if v in vogais:
    contador += 1
print(f"sua palavra tem {contador} vogais")
```
