```python
num1 = int(input("diga um número "))
num2 = int(input("diga outro número "))

soma = num1 + num2
subtracao = num1 - num2
divizao = num1 / num2
multiplicacao = num1 * num2
print(f"{num1} mais {num2} é igual a {soma}, {num1} menos {num2} é igual a {subtracao}, {num1} vezes {num2} é igual a {multiplicacao} e {num1} dividido por {num2} é igual a {divizao}")
```
